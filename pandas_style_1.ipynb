```python
import pandas as pd
import numpy as np
```

```python
random_data = np.random.randint(1, 100, size=(10, 4))
df = pd.DataFrame(random_data, columns=['A', 'B', 'C', 'D'])
nan_indexes = [(0, 1), (2, 3), (4, 0), (5, 2), (7, 3)]

for i, j in nan_indexes:
    df.iat[i, j] = np.nan

print("Original DataFrame:")
display(df)

#additional dependency: Jinja2
def highlight_nan(s):
    return ['background-color: #009900' 
                if pd.isna(v) 
                else '' for v in s
            ]

print("Colored DataFrame with null value highlighted")
styled_df = df.style.apply(highlight_nan)
styled_df
```
